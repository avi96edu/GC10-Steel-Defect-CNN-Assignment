```python
import os
import json
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_recall_fscore_support

warnings.filterwarnings("ignore")
print("Imports loaded successfully.")
from pathlib import Path

# =========================
# CONFIG
# =========================
MODEL_NAME = "mobilenetv3_small"   # "mobilenetv3_small" or "resnet18"
SEED = 42                          # 42, 13, 87

OPTIMIZER_NAME = "adamw"
USE_WEIGHTED_LOSS = True
USE_LABEL_SMOOTHING = True
USE_PRETRAINED = True

NUM_CLASSES = 10

# Fixed coverage checkpoints for reporting
FIXED_COVERAGES = [1.00, 0.95, 0.90, 0.80, 0.70, 0.60, 0.50]

# =========================
# PATHS
# =========================
CANDIDATE_BASE_DIRS = [
    Path("/mnt/e/CVPR"),
    Path("/mnt/d/CVPR"),
    Path(r"E:\CVPR"),
    Path(r"D:\CVPR"),
]

BASE_DIR = None
for p in CANDIDATE_BASE_DIRS:
    if p.exists():
        BASE_DIR = p
        break

if BASE_DIR is None:
    raise FileNotFoundError("Could not find CVPR folder.")

PROJECT_ROOT = BASE_DIR / "steel_failure_aware"
CLOSED_SET_DIR = PROJECT_ROOT / "runs" / "closed_set"
CALIBRATION_DIR = PROJECT_ROOT / "runs" / "calibration"
SELECTIVE_DIR = PROJECT_ROOT / "runs" / "selective_prediction"
SELECTIVE_DIR.mkdir(parents=True, exist_ok=True)

RUN_NAME = (
    f"{MODEL_NAME}__{OPTIMIZER_NAME}"
    f"__weightedCE_{USE_WEIGHTED_LOSS}"
    f"__labelsmooth_{USE_LABEL_SMOOTHING}"
    f"__pretrained_{USE_PRETRAINED}"
    f"__seed_{SEED}"
)

RAW_RUN_DIR = CLOSED_SET_DIR / RUN_NAME
CAL_RUN_DIR = CALIBRATION_DIR / RUN_NAME
OUT_RUN_DIR = SELECTIVE_DIR / RUN_NAME
OUT_RUN_DIR.mkdir(parents=True, exist_ok=True)

RAW_TEST_CSV = RAW_RUN_DIR / "test_predictions.csv"
CAL_TEST_CSV = CAL_RUN_DIR / "test_predictions_calibrated.csv"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_TEST_CSV:", RAW_TEST_CSV, "->", RAW_TEST_CSV.exists())
print("CAL_TEST_CSV:", CAL_TEST_CSV, "->", CAL_TEST_CSV.exists())
print("OUT_RUN_DIR :", OUT_RUN_DIR)
raw_df = pd.read_csv(RAW_TEST_CSV)
cal_df = pd.read_csv(CAL_TEST_CSV)

print("raw_df shape:", raw_df.shape)
print("cal_df shape:", cal_df.shape)

display(raw_df.head(2))
display(cal_df.head(2))
raw_prob_cols = sorted([c for c in raw_df.columns if c.startswith("prob_class_")], key=lambda x: int(x.split("_")[-1]))
raw_logit_cols = sorted([c for c in raw_df.columns if c.startswith("logit_class_")], key=lambda x: int(x.split("_")[-1]))

cal_prob_cols = sorted([c for c in cal_df.columns if c.startswith("cal_prob_class_")], key=lambda x: int(x.split("_")[-1]))
cal_logit_cols = sorted([c for c in cal_df.columns if c.startswith("cal_logit_class_")], key=lambda x: int(x.split("_")[-1]))

assert len(raw_prob_cols) == NUM_CLASSES
assert len(raw_logit_cols) == NUM_CLASSES
assert len(cal_prob_cols) == NUM_CLASSES
assert len(cal_logit_cols) == NUM_CLASSES

y_true = raw_df["y_true"].values.astype(np.int64)
y_pred = raw_df["y_pred"].values.astype(np.int64)

raw_probs = raw_df[raw_prob_cols].values.astype(np.float32)
raw_logits = raw_df[raw_logit_cols].values.astype(np.float32)

cal_probs = cal_df[cal_prob_cols].values.astype(np.float32)
cal_logits = cal_df[cal_logit_cols].values.astype(np.float32)

print("raw_probs :", raw_probs.shape)
print("cal_probs :", cal_probs.shape)
print("raw_logits:", raw_logits.shape)
print("cal_logits:", cal_logits.shape)
print("y_true    :", y_true.shape)
def normalized_entropy(probs: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    ent = -np.sum(probs * np.log(np.clip(probs, eps, 1.0)), axis=1)
    ent = ent / np.log(probs.shape[1])
    return ent

# 1) Raw MSP
raw_msp = np.max(raw_probs, axis=1)

# 2) Calibrated MSP
cal_msp = np.max(cal_probs, axis=1)

# 3) Entropy confidence (higher = better / more confident)
raw_entropy = normalized_entropy(raw_probs)
entropy_conf = 1.0 - raw_entropy

# 4) Max logit
max_logit = np.max(raw_logits, axis=1)

score_dict = {
    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy": entropy_conf,     # confidence form
    "max_logit": max_logit,
}

for k, v in score_dict.items():
    print(k, v.shape, float(v.min()), float(v.max()))
def macro_f1_score_np(y_true_sel, y_pred_sel):
    _, _, macro_f1, _ = precision_recall_fscore_support(
        y_true_sel, y_pred_sel, average="macro", zero_division=0
    )
    return float(macro_f1)

def build_risk_coverage_curve(y_true, y_pred, scores):
    order = np.argsort(-scores)  # descending confidence
    y_true_sorted = y_true[order]
    y_pred_sorted = y_pred[order]
    scores_sorted = scores[order]

    correct = (y_true_sorted == y_pred_sorted).astype(np.float32)
    error = 1.0 - correct

    n = len(y_true_sorted)
    coverages = np.arange(1, n + 1) / n

    cum_correct = np.cumsum(correct)
    cum_error = np.cumsum(error)

    selective_accuracy = cum_correct / np.arange(1, n + 1)
    risk = cum_error / np.arange(1, n + 1)

    curve_df = pd.DataFrame({
        "rank": np.arange(1, n + 1),
        "coverage": coverages,
        "score": scores_sorted,
        "y_true": y_true_sorted,
        "y_pred": y_pred_sorted,
        "correct": correct,
        "risk": risk,
        "selective_accuracy": selective_accuracy,
    })
    return curve_df

def compute_aurc(curve_df):
    # discrete AURC
    return float(curve_df["risk"].mean())

def compute_eaurc(curve_df):
    aurc = compute_aurc(curve_df)
    overall_risk = float(curve_df["risk"].iloc[-1])  # full coverage risk
    eps = 1e-12
    if overall_risk >= 1.0:
        optimal_aurc = 1.0
    else:
        optimal_aurc = overall_risk + (1.0 - overall_risk) * np.log(np.clip(1.0 - overall_risk, eps, 1.0))
    eaurc = aurc - optimal_aurc
    return float(eaurc)

def selective_metrics_at_coverages(y_true, y_pred, scores, coverages):
    order = np.argsort(-scores)
    n = len(y_true)

    rows = []
    for cov in coverages:
        k = max(1, int(np.ceil(cov * n)))
        idx = order[:k]

        y_true_sel = y_true[idx]
        y_pred_sel = y_pred[idx]

        acc = accuracy_score(y_true_sel, y_pred_sel)
        macro_f1 = macro_f1_score_np(y_true_sel, y_pred_sel)

        rows.append({
            "coverage": float(cov),
            "n_selected": int(k),
            "selective_accuracy": float(acc),
            "selective_macro_f1": float(macro_f1),
        })

    return pd.DataFrame(rows)
curve_tables = {}
summary_rows = []
fixed_tables = {}

for score_name, scores in score_dict.items():
    curve_df = build_risk_coverage_curve(y_true, y_pred, scores)
    aurc = compute_aurc(curve_df)
    eaurc = compute_eaurc(curve_df)
    fixed_df = selective_metrics_at_coverages(y_true, y_pred, scores, FIXED_COVERAGES)

    curve_tables[score_name] = curve_df
    fixed_tables[score_name] = fixed_df

    summary_rows.append({
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
        "aurc": aurc,
        "eaurc": eaurc,
        "full_coverage_accuracy": float(curve_df["selective_accuracy"].iloc[-1]),
        "full_coverage_risk": float(curve_df["risk"].iloc[-1]),
    })

summary_df = pd.DataFrame(summary_rows).sort_values("score_name").reset_index(drop=True)
display(summary_df)
for score_name, curve_df in curve_tables.items():
    curve_path = OUT_RUN_DIR / f"risk_coverage_{score_name}.csv"
    curve_df.to_csv(curve_path, index=False)

for score_name, fixed_df in fixed_tables.items():
    fixed_path = OUT_RUN_DIR / f"selective_metrics_fixed_coverage_{score_name}.csv"
    fixed_df.to_csv(fixed_path, index=False)

summary_csv = OUT_RUN_DIR / "selective_prediction_summary.csv"
summary_json = OUT_RUN_DIR / "selective_prediction_summary.json"

summary_df.to_csv(summary_csv, index=False)
summary_df.to_json(summary_json, orient="records", indent=2)

print("Saved:", summary_csv)
print("Saved:", summary_json)
print("Saved risk-coverage and fixed-coverage CSV files.")
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["risk"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Risk")
plt.title(f"Risk-Coverage Curve - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "risk_coverage_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["selective_accuracy"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Accuracy")
plt.title(f"Selective Accuracy vs Coverage - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_accuracy_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, fixed_df in fixed_tables.items():
    plt.plot(fixed_df["coverage"], fixed_df["selective_macro_f1"], marker="o", label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Macro-F1")
plt.title(f"Selective Macro-F1 at Fixed Coverages - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_macro_f1_fixed_coverages.png", dpi=200)
plt.show()
decision_df = pd.DataFrame({
    "filepath": raw_df["filepath"],
    "y_true": y_true,
    "y_pred": y_pred,
    "correct": (y_true == y_pred).astype(int),

    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy_conf": entropy_conf,
    "raw_entropy": raw_entropy,
    "max_logit": max_logit,
})

for score_name, scores in score_dict.items():
    order = np.argsort(-scores)
    rank = np.empty_like(order)
    rank[order] = np.arange(1, len(scores) + 1)
    decision_df[f"rank_{score_name}"] = rank

    for cov in FIXED_COVERAGES:
        k = max(1, int(np.ceil(cov * len(scores))))
        accepted = np.zeros(len(scores), dtype=int)
        accepted[order[:k]] = 1
        cov_tag = str(cov).replace(".", "")
        decision_df[f"accept_{score_name}_cov_{cov_tag}"] = accepted

decision_csv = OUT_RUN_DIR / "confidence_scores_and_decisions.csv"
decision_df.to_csv(decision_csv, index=False)

print("Saved:", decision_csv)
display(decision_df.head())
wide_rows = []

for score_name, fixed_df in fixed_tables.items():
    row = {
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
    }
    for _, r in fixed_df.iterrows():
        cov = str(r["coverage"]).replace(".", "")
        row[f"acc_cov_{cov}"] = float(r["selective_accuracy"])
        row[f"macrof1_cov_{cov}"] = float(r["selective_macro_f1"])
    wide_rows.append(row)

wide_df = pd.DataFrame(wide_rows)
wide_csv = OUT_RUN_DIR / "selective_metrics_fixed_coverage_wide.csv"
wide_df.to_csv(wide_csv, index=False)

print("Saved:", wide_csv)
display(wide_df)
print("\nMODEL:", MODEL_NAME)
print("SEED :", SEED)

print("\nAURC / E-AURC summary")
display(summary_df)

for score_name, fixed_df in fixed_tables.items():
    print(f"\nFixed coverage metrics -> {score_name}")
    display(fixed_df)
saved_files = sorted([p.name for p in OUT_RUN_DIR.iterdir()])
print("Saved files:")
for f in saved_files:
    print("-", f)
```

```python
import os
import json
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_recall_fscore_support

warnings.filterwarnings("ignore")
print("Imports loaded successfully.")
from pathlib import Path

# =========================
# CONFIG
# =========================
MODEL_NAME = "mobilenetv3_small"   # "mobilenetv3_small" or "resnet18"
SEED = 87                          # 42, 13, 87

OPTIMIZER_NAME = "adamw"
USE_WEIGHTED_LOSS = True
USE_LABEL_SMOOTHING = True
USE_PRETRAINED = True

NUM_CLASSES = 10

# Fixed coverage checkpoints for reporting
FIXED_COVERAGES = [1.00, 0.95, 0.90, 0.80, 0.70, 0.60, 0.50]

# =========================
# PATHS
# =========================
CANDIDATE_BASE_DIRS = [
    Path("/mnt/e/CVPR"),
    Path("/mnt/d/CVPR"),
    Path(r"E:\CVPR"),
    Path(r"D:\CVPR"),
]

BASE_DIR = None
for p in CANDIDATE_BASE_DIRS:
    if p.exists():
        BASE_DIR = p
        break

if BASE_DIR is None:
    raise FileNotFoundError("Could not find CVPR folder.")

PROJECT_ROOT = BASE_DIR / "steel_failure_aware"
CLOSED_SET_DIR = PROJECT_ROOT / "runs" / "closed_set"
CALIBRATION_DIR = PROJECT_ROOT / "runs" / "calibration"
SELECTIVE_DIR = PROJECT_ROOT / "runs" / "selective_prediction"
SELECTIVE_DIR.mkdir(parents=True, exist_ok=True)

RUN_NAME = (
    f"{MODEL_NAME}__{OPTIMIZER_NAME}"
    f"__weightedCE_{USE_WEIGHTED_LOSS}"
    f"__labelsmooth_{USE_LABEL_SMOOTHING}"
    f"__pretrained_{USE_PRETRAINED}"
    f"__seed_{SEED}"
)

RAW_RUN_DIR = CLOSED_SET_DIR / RUN_NAME
CAL_RUN_DIR = CALIBRATION_DIR / RUN_NAME
OUT_RUN_DIR = SELECTIVE_DIR / RUN_NAME
OUT_RUN_DIR.mkdir(parents=True, exist_ok=True)

RAW_TEST_CSV = RAW_RUN_DIR / "test_predictions.csv"
CAL_TEST_CSV = CAL_RUN_DIR / "test_predictions_calibrated.csv"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_TEST_CSV:", RAW_TEST_CSV, "->", RAW_TEST_CSV.exists())
print("CAL_TEST_CSV:", CAL_TEST_CSV, "->", CAL_TEST_CSV.exists())
print("OUT_RUN_DIR :", OUT_RUN_DIR)
raw_df = pd.read_csv(RAW_TEST_CSV)
cal_df = pd.read_csv(CAL_TEST_CSV)

print("raw_df shape:", raw_df.shape)
print("cal_df shape:", cal_df.shape)

display(raw_df.head(2))
display(cal_df.head(2))
raw_prob_cols = sorted([c for c in raw_df.columns if c.startswith("prob_class_")], key=lambda x: int(x.split("_")[-1]))
raw_logit_cols = sorted([c for c in raw_df.columns if c.startswith("logit_class_")], key=lambda x: int(x.split("_")[-1]))

cal_prob_cols = sorted([c for c in cal_df.columns if c.startswith("cal_prob_class_")], key=lambda x: int(x.split("_")[-1]))
cal_logit_cols = sorted([c for c in cal_df.columns if c.startswith("cal_logit_class_")], key=lambda x: int(x.split("_")[-1]))

assert len(raw_prob_cols) == NUM_CLASSES
assert len(raw_logit_cols) == NUM_CLASSES
assert len(cal_prob_cols) == NUM_CLASSES
assert len(cal_logit_cols) == NUM_CLASSES

y_true = raw_df["y_true"].values.astype(np.int64)
y_pred = raw_df["y_pred"].values.astype(np.int64)

raw_probs = raw_df[raw_prob_cols].values.astype(np.float32)
raw_logits = raw_df[raw_logit_cols].values.astype(np.float32)

cal_probs = cal_df[cal_prob_cols].values.astype(np.float32)
cal_logits = cal_df[cal_logit_cols].values.astype(np.float32)

print("raw_probs :", raw_probs.shape)
print("cal_probs :", cal_probs.shape)
print("raw_logits:", raw_logits.shape)
print("cal_logits:", cal_logits.shape)
print("y_true    :", y_true.shape)
def normalized_entropy(probs: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    ent = -np.sum(probs * np.log(np.clip(probs, eps, 1.0)), axis=1)
    ent = ent / np.log(probs.shape[1])
    return ent

# 1) Raw MSP
raw_msp = np.max(raw_probs, axis=1)

# 2) Calibrated MSP
cal_msp = np.max(cal_probs, axis=1)

# 3) Entropy confidence (higher = better / more confident)
raw_entropy = normalized_entropy(raw_probs)
entropy_conf = 1.0 - raw_entropy

# 4) Max logit
max_logit = np.max(raw_logits, axis=1)

score_dict = {
    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy": entropy_conf,     # confidence form
    "max_logit": max_logit,
}

for k, v in score_dict.items():
    print(k, v.shape, float(v.min()), float(v.max()))
def macro_f1_score_np(y_true_sel, y_pred_sel):
    _, _, macro_f1, _ = precision_recall_fscore_support(
        y_true_sel, y_pred_sel, average="macro", zero_division=0
    )
    return float(macro_f1)

def build_risk_coverage_curve(y_true, y_pred, scores):
    order = np.argsort(-scores)  # descending confidence
    y_true_sorted = y_true[order]
    y_pred_sorted = y_pred[order]
    scores_sorted = scores[order]

    correct = (y_true_sorted == y_pred_sorted).astype(np.float32)
    error = 1.0 - correct

    n = len(y_true_sorted)
    coverages = np.arange(1, n + 1) / n

    cum_correct = np.cumsum(correct)
    cum_error = np.cumsum(error)

    selective_accuracy = cum_correct / np.arange(1, n + 1)
    risk = cum_error / np.arange(1, n + 1)

    curve_df = pd.DataFrame({
        "rank": np.arange(1, n + 1),
        "coverage": coverages,
        "score": scores_sorted,
        "y_true": y_true_sorted,
        "y_pred": y_pred_sorted,
        "correct": correct,
        "risk": risk,
        "selective_accuracy": selective_accuracy,
    })
    return curve_df

def compute_aurc(curve_df):
    # discrete AURC
    return float(curve_df["risk"].mean())

def compute_eaurc(curve_df):
    aurc = compute_aurc(curve_df)
    overall_risk = float(curve_df["risk"].iloc[-1])  # full coverage risk
    eps = 1e-12
    if overall_risk >= 1.0:
        optimal_aurc = 1.0
    else:
        optimal_aurc = overall_risk + (1.0 - overall_risk) * np.log(np.clip(1.0 - overall_risk, eps, 1.0))
    eaurc = aurc - optimal_aurc
    return float(eaurc)

def selective_metrics_at_coverages(y_true, y_pred, scores, coverages):
    order = np.argsort(-scores)
    n = len(y_true)

    rows = []
    for cov in coverages:
        k = max(1, int(np.ceil(cov * n)))
        idx = order[:k]

        y_true_sel = y_true[idx]
        y_pred_sel = y_pred[idx]

        acc = accuracy_score(y_true_sel, y_pred_sel)
        macro_f1 = macro_f1_score_np(y_true_sel, y_pred_sel)

        rows.append({
            "coverage": float(cov),
            "n_selected": int(k),
            "selective_accuracy": float(acc),
            "selective_macro_f1": float(macro_f1),
        })

    return pd.DataFrame(rows)
curve_tables = {}
summary_rows = []
fixed_tables = {}

for score_name, scores in score_dict.items():
    curve_df = build_risk_coverage_curve(y_true, y_pred, scores)
    aurc = compute_aurc(curve_df)
    eaurc = compute_eaurc(curve_df)
    fixed_df = selective_metrics_at_coverages(y_true, y_pred, scores, FIXED_COVERAGES)

    curve_tables[score_name] = curve_df
    fixed_tables[score_name] = fixed_df

    summary_rows.append({
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
        "aurc": aurc,
        "eaurc": eaurc,
        "full_coverage_accuracy": float(curve_df["selective_accuracy"].iloc[-1]),
        "full_coverage_risk": float(curve_df["risk"].iloc[-1]),
    })

summary_df = pd.DataFrame(summary_rows).sort_values("score_name").reset_index(drop=True)
display(summary_df)
for score_name, curve_df in curve_tables.items():
    curve_path = OUT_RUN_DIR / f"risk_coverage_{score_name}.csv"
    curve_df.to_csv(curve_path, index=False)

for score_name, fixed_df in fixed_tables.items():
    fixed_path = OUT_RUN_DIR / f"selective_metrics_fixed_coverage_{score_name}.csv"
    fixed_df.to_csv(fixed_path, index=False)

summary_csv = OUT_RUN_DIR / "selective_prediction_summary.csv"
summary_json = OUT_RUN_DIR / "selective_prediction_summary.json"

summary_df.to_csv(summary_csv, index=False)
summary_df.to_json(summary_json, orient="records", indent=2)

print("Saved:", summary_csv)
print("Saved:", summary_json)
print("Saved risk-coverage and fixed-coverage CSV files.")
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["risk"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Risk")
plt.title(f"Risk-Coverage Curve - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "risk_coverage_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["selective_accuracy"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Accuracy")
plt.title(f"Selective Accuracy vs Coverage - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_accuracy_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, fixed_df in fixed_tables.items():
    plt.plot(fixed_df["coverage"], fixed_df["selective_macro_f1"], marker="o", label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Macro-F1")
plt.title(f"Selective Macro-F1 at Fixed Coverages - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_macro_f1_fixed_coverages.png", dpi=200)
plt.show()
decision_df = pd.DataFrame({
    "filepath": raw_df["filepath"],
    "y_true": y_true,
    "y_pred": y_pred,
    "correct": (y_true == y_pred).astype(int),

    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy_conf": entropy_conf,
    "raw_entropy": raw_entropy,
    "max_logit": max_logit,
})

for score_name, scores in score_dict.items():
    order = np.argsort(-scores)
    rank = np.empty_like(order)
    rank[order] = np.arange(1, len(scores) + 1)
    decision_df[f"rank_{score_name}"] = rank

    for cov in FIXED_COVERAGES:
        k = max(1, int(np.ceil(cov * len(scores))))
        accepted = np.zeros(len(scores), dtype=int)
        accepted[order[:k]] = 1
        cov_tag = str(cov).replace(".", "")
        decision_df[f"accept_{score_name}_cov_{cov_tag}"] = accepted

decision_csv = OUT_RUN_DIR / "confidence_scores_and_decisions.csv"
decision_df.to_csv(decision_csv, index=False)

print("Saved:", decision_csv)
display(decision_df.head())
wide_rows = []

for score_name, fixed_df in fixed_tables.items():
    row = {
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
    }
    for _, r in fixed_df.iterrows():
        cov = str(r["coverage"]).replace(".", "")
        row[f"acc_cov_{cov}"] = float(r["selective_accuracy"])
        row[f"macrof1_cov_{cov}"] = float(r["selective_macro_f1"])
    wide_rows.append(row)

wide_df = pd.DataFrame(wide_rows)
wide_csv = OUT_RUN_DIR / "selective_metrics_fixed_coverage_wide.csv"
wide_df.to_csv(wide_csv, index=False)

print("Saved:", wide_csv)
display(wide_df)
print("\nMODEL:", MODEL_NAME)
print("SEED :", SEED)

print("\nAURC / E-AURC summary")
display(summary_df)

for score_name, fixed_df in fixed_tables.items():
    print(f"\nFixed coverage metrics -> {score_name}")
    display(fixed_df)
saved_files = sorted([p.name for p in OUT_RUN_DIR.iterdir()])
print("Saved files:")
for f in saved_files:
    print("-", f)
```

```python
import os
import json
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_recall_fscore_support

warnings.filterwarnings("ignore")
print("Imports loaded successfully.")
from pathlib import Path

# =========================
# CONFIG
# =========================
MODEL_NAME = "resnet18"   # "mobilenetv3_small" or "resnet18"
SEED = 42                          # 42, 13, 87

OPTIMIZER_NAME = "adamw"
USE_WEIGHTED_LOSS = True
USE_LABEL_SMOOTHING = True
USE_PRETRAINED = True

NUM_CLASSES = 10

# Fixed coverage checkpoints for reporting
FIXED_COVERAGES = [1.00, 0.95, 0.90, 0.80, 0.70, 0.60, 0.50]

# =========================
# PATHS
# =========================
CANDIDATE_BASE_DIRS = [
    Path("/mnt/e/CVPR"),
    Path("/mnt/d/CVPR"),
    Path(r"E:\CVPR"),
    Path(r"D:\CVPR"),
]

BASE_DIR = None
for p in CANDIDATE_BASE_DIRS:
    if p.exists():
        BASE_DIR = p
        break

if BASE_DIR is None:
    raise FileNotFoundError("Could not find CVPR folder.")

PROJECT_ROOT = BASE_DIR / "steel_failure_aware"
CLOSED_SET_DIR = PROJECT_ROOT / "runs" / "closed_set"
CALIBRATION_DIR = PROJECT_ROOT / "runs" / "calibration"
SELECTIVE_DIR = PROJECT_ROOT / "runs" / "selective_prediction"
SELECTIVE_DIR.mkdir(parents=True, exist_ok=True)

RUN_NAME = (
    f"{MODEL_NAME}__{OPTIMIZER_NAME}"
    f"__weightedCE_{USE_WEIGHTED_LOSS}"
    f"__labelsmooth_{USE_LABEL_SMOOTHING}"
    f"__pretrained_{USE_PRETRAINED}"
    f"__seed_{SEED}"
)

RAW_RUN_DIR = CLOSED_SET_DIR / RUN_NAME
CAL_RUN_DIR = CALIBRATION_DIR / RUN_NAME
OUT_RUN_DIR = SELECTIVE_DIR / RUN_NAME
OUT_RUN_DIR.mkdir(parents=True, exist_ok=True)

RAW_TEST_CSV = RAW_RUN_DIR / "test_predictions.csv"
CAL_TEST_CSV = CAL_RUN_DIR / "test_predictions_calibrated.csv"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_TEST_CSV:", RAW_TEST_CSV, "->", RAW_TEST_CSV.exists())
print("CAL_TEST_CSV:", CAL_TEST_CSV, "->", CAL_TEST_CSV.exists())
print("OUT_RUN_DIR :", OUT_RUN_DIR)
raw_df = pd.read_csv(RAW_TEST_CSV)
cal_df = pd.read_csv(CAL_TEST_CSV)

print("raw_df shape:", raw_df.shape)
print("cal_df shape:", cal_df.shape)

display(raw_df.head(2))
display(cal_df.head(2))
raw_prob_cols = sorted([c for c in raw_df.columns if c.startswith("prob_class_")], key=lambda x: int(x.split("_")[-1]))
raw_logit_cols = sorted([c for c in raw_df.columns if c.startswith("logit_class_")], key=lambda x: int(x.split("_")[-1]))

cal_prob_cols = sorted([c for c in cal_df.columns if c.startswith("cal_prob_class_")], key=lambda x: int(x.split("_")[-1]))
cal_logit_cols = sorted([c for c in cal_df.columns if c.startswith("cal_logit_class_")], key=lambda x: int(x.split("_")[-1]))

assert len(raw_prob_cols) == NUM_CLASSES
assert len(raw_logit_cols) == NUM_CLASSES
assert len(cal_prob_cols) == NUM_CLASSES
assert len(cal_logit_cols) == NUM_CLASSES

y_true = raw_df["y_true"].values.astype(np.int64)
y_pred = raw_df["y_pred"].values.astype(np.int64)

raw_probs = raw_df[raw_prob_cols].values.astype(np.float32)
raw_logits = raw_df[raw_logit_cols].values.astype(np.float32)

cal_probs = cal_df[cal_prob_cols].values.astype(np.float32)
cal_logits = cal_df[cal_logit_cols].values.astype(np.float32)

print("raw_probs :", raw_probs.shape)
print("cal_probs :", cal_probs.shape)
print("raw_logits:", raw_logits.shape)
print("cal_logits:", cal_logits.shape)
print("y_true    :", y_true.shape)
def normalized_entropy(probs: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    ent = -np.sum(probs * np.log(np.clip(probs, eps, 1.0)), axis=1)
    ent = ent / np.log(probs.shape[1])
    return ent

# 1) Raw MSP
raw_msp = np.max(raw_probs, axis=1)

# 2) Calibrated MSP
cal_msp = np.max(cal_probs, axis=1)

# 3) Entropy confidence (higher = better / more confident)
raw_entropy = normalized_entropy(raw_probs)
entropy_conf = 1.0 - raw_entropy

# 4) Max logit
max_logit = np.max(raw_logits, axis=1)

score_dict = {
    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy": entropy_conf,     # confidence form
    "max_logit": max_logit,
}

for k, v in score_dict.items():
    print(k, v.shape, float(v.min()), float(v.max()))
def macro_f1_score_np(y_true_sel, y_pred_sel):
    _, _, macro_f1, _ = precision_recall_fscore_support(
        y_true_sel, y_pred_sel, average="macro", zero_division=0
    )
    return float(macro_f1)

def build_risk_coverage_curve(y_true, y_pred, scores):
    order = np.argsort(-scores)  # descending confidence
    y_true_sorted = y_true[order]
    y_pred_sorted = y_pred[order]
    scores_sorted = scores[order]

    correct = (y_true_sorted == y_pred_sorted).astype(np.float32)
    error = 1.0 - correct

    n = len(y_true_sorted)
    coverages = np.arange(1, n + 1) / n

    cum_correct = np.cumsum(correct)
    cum_error = np.cumsum(error)

    selective_accuracy = cum_correct / np.arange(1, n + 1)
    risk = cum_error / np.arange(1, n + 1)

    curve_df = pd.DataFrame({
        "rank": np.arange(1, n + 1),
        "coverage": coverages,
        "score": scores_sorted,
        "y_true": y_true_sorted,
        "y_pred": y_pred_sorted,
        "correct": correct,
        "risk": risk,
        "selective_accuracy": selective_accuracy,
    })
    return curve_df

def compute_aurc(curve_df):
    # discrete AURC
    return float(curve_df["risk"].mean())

def compute_eaurc(curve_df):
    aurc = compute_aurc(curve_df)
    overall_risk = float(curve_df["risk"].iloc[-1])  # full coverage risk
    eps = 1e-12
    if overall_risk >= 1.0:
        optimal_aurc = 1.0
    else:
        optimal_aurc = overall_risk + (1.0 - overall_risk) * np.log(np.clip(1.0 - overall_risk, eps, 1.0))
    eaurc = aurc - optimal_aurc
    return float(eaurc)

def selective_metrics_at_coverages(y_true, y_pred, scores, coverages):
    order = np.argsort(-scores)
    n = len(y_true)

    rows = []
    for cov in coverages:
        k = max(1, int(np.ceil(cov * n)))
        idx = order[:k]

        y_true_sel = y_true[idx]
        y_pred_sel = y_pred[idx]

        acc = accuracy_score(y_true_sel, y_pred_sel)
        macro_f1 = macro_f1_score_np(y_true_sel, y_pred_sel)

        rows.append({
            "coverage": float(cov),
            "n_selected": int(k),
            "selective_accuracy": float(acc),
            "selective_macro_f1": float(macro_f1),
        })

    return pd.DataFrame(rows)
curve_tables = {}
summary_rows = []
fixed_tables = {}

for score_name, scores in score_dict.items():
    curve_df = build_risk_coverage_curve(y_true, y_pred, scores)
    aurc = compute_aurc(curve_df)
    eaurc = compute_eaurc(curve_df)
    fixed_df = selective_metrics_at_coverages(y_true, y_pred, scores, FIXED_COVERAGES)

    curve_tables[score_name] = curve_df
    fixed_tables[score_name] = fixed_df

    summary_rows.append({
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
        "aurc": aurc,
        "eaurc": eaurc,
        "full_coverage_accuracy": float(curve_df["selective_accuracy"].iloc[-1]),
        "full_coverage_risk": float(curve_df["risk"].iloc[-1]),
    })

summary_df = pd.DataFrame(summary_rows).sort_values("score_name").reset_index(drop=True)
display(summary_df)
for score_name, curve_df in curve_tables.items():
    curve_path = OUT_RUN_DIR / f"risk_coverage_{score_name}.csv"
    curve_df.to_csv(curve_path, index=False)

for score_name, fixed_df in fixed_tables.items():
    fixed_path = OUT_RUN_DIR / f"selective_metrics_fixed_coverage_{score_name}.csv"
    fixed_df.to_csv(fixed_path, index=False)

summary_csv = OUT_RUN_DIR / "selective_prediction_summary.csv"
summary_json = OUT_RUN_DIR / "selective_prediction_summary.json"

summary_df.to_csv(summary_csv, index=False)
summary_df.to_json(summary_json, orient="records", indent=2)

print("Saved:", summary_csv)
print("Saved:", summary_json)
print("Saved risk-coverage and fixed-coverage CSV files.")
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["risk"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Risk")
plt.title(f"Risk-Coverage Curve - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "risk_coverage_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["selective_accuracy"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Accuracy")
plt.title(f"Selective Accuracy vs Coverage - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_accuracy_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, fixed_df in fixed_tables.items():
    plt.plot(fixed_df["coverage"], fixed_df["selective_macro_f1"], marker="o", label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Macro-F1")
plt.title(f"Selective Macro-F1 at Fixed Coverages - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_macro_f1_fixed_coverages.png", dpi=200)
plt.show()
decision_df = pd.DataFrame({
    "filepath": raw_df["filepath"],
    "y_true": y_true,
    "y_pred": y_pred,
    "correct": (y_true == y_pred).astype(int),

    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy_conf": entropy_conf,
    "raw_entropy": raw_entropy,
    "max_logit": max_logit,
})

for score_name, scores in score_dict.items():
    order = np.argsort(-scores)
    rank = np.empty_like(order)
    rank[order] = np.arange(1, len(scores) + 1)
    decision_df[f"rank_{score_name}"] = rank

    for cov in FIXED_COVERAGES:
        k = max(1, int(np.ceil(cov * len(scores))))
        accepted = np.zeros(len(scores), dtype=int)
        accepted[order[:k]] = 1
        cov_tag = str(cov).replace(".", "")
        decision_df[f"accept_{score_name}_cov_{cov_tag}"] = accepted

decision_csv = OUT_RUN_DIR / "confidence_scores_and_decisions.csv"
decision_df.to_csv(decision_csv, index=False)

print("Saved:", decision_csv)
display(decision_df.head())
wide_rows = []

for score_name, fixed_df in fixed_tables.items():
    row = {
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
    }
    for _, r in fixed_df.iterrows():
        cov = str(r["coverage"]).replace(".", "")
        row[f"acc_cov_{cov}"] = float(r["selective_accuracy"])
        row[f"macrof1_cov_{cov}"] = float(r["selective_macro_f1"])
    wide_rows.append(row)

wide_df = pd.DataFrame(wide_rows)
wide_csv = OUT_RUN_DIR / "selective_metrics_fixed_coverage_wide.csv"
wide_df.to_csv(wide_csv, index=False)

print("Saved:", wide_csv)
display(wide_df)
print("\nMODEL:", MODEL_NAME)
print("SEED :", SEED)

print("\nAURC / E-AURC summary")
display(summary_df)

for score_name, fixed_df in fixed_tables.items():
    print(f"\nFixed coverage metrics -> {score_name}")
    display(fixed_df)
saved_files = sorted([p.name for p in OUT_RUN_DIR.iterdir()])
print("Saved files:")
for f in saved_files:
    print("-", f)
```

```python
import os
import json
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, precision_recall_fscore_support

warnings.filterwarnings("ignore")
print("Imports loaded successfully.")
from pathlib import Path

# =========================
# CONFIG
# =========================
MODEL_NAME = "resnet18"   # "mobilenetv3_small" or "resnet18"
SEED = 87                          # 42, 13, 87

OPTIMIZER_NAME = "adamw"
USE_WEIGHTED_LOSS = True
USE_LABEL_SMOOTHING = True
USE_PRETRAINED = True

NUM_CLASSES = 10

# Fixed coverage checkpoints for reporting
FIXED_COVERAGES = [1.00, 0.95, 0.90, 0.80, 0.70, 0.60, 0.50]

# =========================
# PATHS
# =========================
CANDIDATE_BASE_DIRS = [
    Path("/mnt/e/CVPR"),
    Path("/mnt/d/CVPR"),
    Path(r"E:\CVPR"),
    Path(r"D:\CVPR"),
]

BASE_DIR = None
for p in CANDIDATE_BASE_DIRS:
    if p.exists():
        BASE_DIR = p
        break

if BASE_DIR is None:
    raise FileNotFoundError("Could not find CVPR folder.")

PROJECT_ROOT = BASE_DIR / "steel_failure_aware"
CLOSED_SET_DIR = PROJECT_ROOT / "runs" / "closed_set"
CALIBRATION_DIR = PROJECT_ROOT / "runs" / "calibration"
SELECTIVE_DIR = PROJECT_ROOT / "runs" / "selective_prediction"
SELECTIVE_DIR.mkdir(parents=True, exist_ok=True)

RUN_NAME = (
    f"{MODEL_NAME}__{OPTIMIZER_NAME}"
    f"__weightedCE_{USE_WEIGHTED_LOSS}"
    f"__labelsmooth_{USE_LABEL_SMOOTHING}"
    f"__pretrained_{USE_PRETRAINED}"
    f"__seed_{SEED}"
)

RAW_RUN_DIR = CLOSED_SET_DIR / RUN_NAME
CAL_RUN_DIR = CALIBRATION_DIR / RUN_NAME
OUT_RUN_DIR = SELECTIVE_DIR / RUN_NAME
OUT_RUN_DIR.mkdir(parents=True, exist_ok=True)

RAW_TEST_CSV = RAW_RUN_DIR / "test_predictions.csv"
CAL_TEST_CSV = CAL_RUN_DIR / "test_predictions_calibrated.csv"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("RAW_TEST_CSV:", RAW_TEST_CSV, "->", RAW_TEST_CSV.exists())
print("CAL_TEST_CSV:", CAL_TEST_CSV, "->", CAL_TEST_CSV.exists())
print("OUT_RUN_DIR :", OUT_RUN_DIR)
raw_df = pd.read_csv(RAW_TEST_CSV)
cal_df = pd.read_csv(CAL_TEST_CSV)

print("raw_df shape:", raw_df.shape)
print("cal_df shape:", cal_df.shape)

display(raw_df.head(2))
display(cal_df.head(2))
raw_prob_cols = sorted([c for c in raw_df.columns if c.startswith("prob_class_")], key=lambda x: int(x.split("_")[-1]))
raw_logit_cols = sorted([c for c in raw_df.columns if c.startswith("logit_class_")], key=lambda x: int(x.split("_")[-1]))

cal_prob_cols = sorted([c for c in cal_df.columns if c.startswith("cal_prob_class_")], key=lambda x: int(x.split("_")[-1]))
cal_logit_cols = sorted([c for c in cal_df.columns if c.startswith("cal_logit_class_")], key=lambda x: int(x.split("_")[-1]))

assert len(raw_prob_cols) == NUM_CLASSES
assert len(raw_logit_cols) == NUM_CLASSES
assert len(cal_prob_cols) == NUM_CLASSES
assert len(cal_logit_cols) == NUM_CLASSES

y_true = raw_df["y_true"].values.astype(np.int64)
y_pred = raw_df["y_pred"].values.astype(np.int64)

raw_probs = raw_df[raw_prob_cols].values.astype(np.float32)
raw_logits = raw_df[raw_logit_cols].values.astype(np.float32)

cal_probs = cal_df[cal_prob_cols].values.astype(np.float32)
cal_logits = cal_df[cal_logit_cols].values.astype(np.float32)

print("raw_probs :", raw_probs.shape)
print("cal_probs :", cal_probs.shape)
print("raw_logits:", raw_logits.shape)
print("cal_logits:", cal_logits.shape)
print("y_true    :", y_true.shape)
def normalized_entropy(probs: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    ent = -np.sum(probs * np.log(np.clip(probs, eps, 1.0)), axis=1)
    ent = ent / np.log(probs.shape[1])
    return ent

# 1) Raw MSP
raw_msp = np.max(raw_probs, axis=1)

# 2) Calibrated MSP
cal_msp = np.max(cal_probs, axis=1)

# 3) Entropy confidence (higher = better / more confident)
raw_entropy = normalized_entropy(raw_probs)
entropy_conf = 1.0 - raw_entropy

# 4) Max logit
max_logit = np.max(raw_logits, axis=1)

score_dict = {
    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy": entropy_conf,     # confidence form
    "max_logit": max_logit,
}

for k, v in score_dict.items():
    print(k, v.shape, float(v.min()), float(v.max()))
def macro_f1_score_np(y_true_sel, y_pred_sel):
    _, _, macro_f1, _ = precision_recall_fscore_support(
        y_true_sel, y_pred_sel, average="macro", zero_division=0
    )
    return float(macro_f1)

def build_risk_coverage_curve(y_true, y_pred, scores):
    order = np.argsort(-scores)  # descending confidence
    y_true_sorted = y_true[order]
    y_pred_sorted = y_pred[order]
    scores_sorted = scores[order]

    correct = (y_true_sorted == y_pred_sorted).astype(np.float32)
    error = 1.0 - correct

    n = len(y_true_sorted)
    coverages = np.arange(1, n + 1) / n

    cum_correct = np.cumsum(correct)
    cum_error = np.cumsum(error)

    selective_accuracy = cum_correct / np.arange(1, n + 1)
    risk = cum_error / np.arange(1, n + 1)

    curve_df = pd.DataFrame({
        "rank": np.arange(1, n + 1),
        "coverage": coverages,
        "score": scores_sorted,
        "y_true": y_true_sorted,
        "y_pred": y_pred_sorted,
        "correct": correct,
        "risk": risk,
        "selective_accuracy": selective_accuracy,
    })
    return curve_df

def compute_aurc(curve_df):
    # discrete AURC
    return float(curve_df["risk"].mean())

def compute_eaurc(curve_df):
    aurc = compute_aurc(curve_df)
    overall_risk = float(curve_df["risk"].iloc[-1])  # full coverage risk
    eps = 1e-12
    if overall_risk >= 1.0:
        optimal_aurc = 1.0
    else:
        optimal_aurc = overall_risk + (1.0 - overall_risk) * np.log(np.clip(1.0 - overall_risk, eps, 1.0))
    eaurc = aurc - optimal_aurc
    return float(eaurc)

def selective_metrics_at_coverages(y_true, y_pred, scores, coverages):
    order = np.argsort(-scores)
    n = len(y_true)

    rows = []
    for cov in coverages:
        k = max(1, int(np.ceil(cov * n)))
        idx = order[:k]

        y_true_sel = y_true[idx]
        y_pred_sel = y_pred[idx]

        acc = accuracy_score(y_true_sel, y_pred_sel)
        macro_f1 = macro_f1_score_np(y_true_sel, y_pred_sel)

        rows.append({
            "coverage": float(cov),
            "n_selected": int(k),
            "selective_accuracy": float(acc),
            "selective_macro_f1": float(macro_f1),
        })

    return pd.DataFrame(rows)
curve_tables = {}
summary_rows = []
fixed_tables = {}

for score_name, scores in score_dict.items():
    curve_df = build_risk_coverage_curve(y_true, y_pred, scores)
    aurc = compute_aurc(curve_df)
    eaurc = compute_eaurc(curve_df)
    fixed_df = selective_metrics_at_coverages(y_true, y_pred, scores, FIXED_COVERAGES)

    curve_tables[score_name] = curve_df
    fixed_tables[score_name] = fixed_df

    summary_rows.append({
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
        "aurc": aurc,
        "eaurc": eaurc,
        "full_coverage_accuracy": float(curve_df["selective_accuracy"].iloc[-1]),
        "full_coverage_risk": float(curve_df["risk"].iloc[-1]),
    })

summary_df = pd.DataFrame(summary_rows).sort_values("score_name").reset_index(drop=True)
display(summary_df)
for score_name, curve_df in curve_tables.items():
    curve_path = OUT_RUN_DIR / f"risk_coverage_{score_name}.csv"
    curve_df.to_csv(curve_path, index=False)

for score_name, fixed_df in fixed_tables.items():
    fixed_path = OUT_RUN_DIR / f"selective_metrics_fixed_coverage_{score_name}.csv"
    fixed_df.to_csv(fixed_path, index=False)

summary_csv = OUT_RUN_DIR / "selective_prediction_summary.csv"
summary_json = OUT_RUN_DIR / "selective_prediction_summary.json"

summary_df.to_csv(summary_csv, index=False)
summary_df.to_json(summary_json, orient="records", indent=2)

print("Saved:", summary_csv)
print("Saved:", summary_json)
print("Saved risk-coverage and fixed-coverage CSV files.")
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["risk"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Risk")
plt.title(f"Risk-Coverage Curve - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "risk_coverage_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, curve_df in curve_tables.items():
    plt.plot(curve_df["coverage"], curve_df["selective_accuracy"], label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Accuracy")
plt.title(f"Selective Accuracy vs Coverage - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_accuracy_all_scores.png", dpi=200)
plt.show()
plt.figure(figsize=(8, 6))

for score_name, fixed_df in fixed_tables.items():
    plt.plot(fixed_df["coverage"], fixed_df["selective_macro_f1"], marker="o", label=score_name)

plt.xlabel("Coverage")
plt.ylabel("Selective Macro-F1")
plt.title(f"Selective Macro-F1 at Fixed Coverages - {MODEL_NAME} - seed {SEED}")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUT_RUN_DIR / "selective_macro_f1_fixed_coverages.png", dpi=200)
plt.show()
decision_df = pd.DataFrame({
    "filepath": raw_df["filepath"],
    "y_true": y_true,
    "y_pred": y_pred,
    "correct": (y_true == y_pred).astype(int),

    "raw_msp": raw_msp,
    "calibrated_msp": cal_msp,
    "entropy_conf": entropy_conf,
    "raw_entropy": raw_entropy,
    "max_logit": max_logit,
})

for score_name, scores in score_dict.items():
    order = np.argsort(-scores)
    rank = np.empty_like(order)
    rank[order] = np.arange(1, len(scores) + 1)
    decision_df[f"rank_{score_name}"] = rank

    for cov in FIXED_COVERAGES:
        k = max(1, int(np.ceil(cov * len(scores))))
        accepted = np.zeros(len(scores), dtype=int)
        accepted[order[:k]] = 1
        cov_tag = str(cov).replace(".", "")
        decision_df[f"accept_{score_name}_cov_{cov_tag}"] = accepted

decision_csv = OUT_RUN_DIR / "confidence_scores_and_decisions.csv"
decision_df.to_csv(decision_csv, index=False)

print("Saved:", decision_csv)
display(decision_df.head())
wide_rows = []

for score_name, fixed_df in fixed_tables.items():
    row = {
        "model": MODEL_NAME,
        "seed": SEED,
        "score_name": score_name,
    }
    for _, r in fixed_df.iterrows():
        cov = str(r["coverage"]).replace(".", "")
        row[f"acc_cov_{cov}"] = float(r["selective_accuracy"])
        row[f"macrof1_cov_{cov}"] = float(r["selective_macro_f1"])
    wide_rows.append(row)

wide_df = pd.DataFrame(wide_rows)
wide_csv = OUT_RUN_DIR / "selective_metrics_fixed_coverage_wide.csv"
wide_df.to_csv(wide_csv, index=False)

print("Saved:", wide_csv)
display(wide_df)
print("\nMODEL:", MODEL_NAME)
print("SEED :", SEED)

print("\nAURC / E-AURC summary")
display(summary_df)

for score_name, fixed_df in fixed_tables.items():
    print(f"\nFixed coverage metrics -> {score_name}")
    display(fixed_df)
saved_files = sorted([p.name for p in OUT_RUN_DIR.iterdir()])
print("Saved files:")
for f in saved_files:
    print("-", f)
```
